Scraping The Bestselling Electronic product from Amazon

```python
no_pages = 2

def get_data(pageNo):  
    headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:66.0) Gecko/20100101 Firefox/66.0", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

    r = requests.get('https://www.amazon.in/gp/bestsellers/electronics/ref=zg_bs_pg_'+str(pageNo)+'?ie=UTF8&pg='+str(pageNo), headers=headers)#, proxies=proxies)
    content = r.content
    soup = BeautifulSoup(content)
    #print(soup)

    alls = []
    for d in soup.findAll('div', attrs={'class':'a-section a-spacing-none aok-relative'}):
        #print(d)
        name = d.find('span', attrs={'class':'zg-text-center-align'})
        n = name.find_all('img', alt=True)
        #print(n[0]['alt'])
        rating = d.find('span', attrs={'class':'a-icon-alt'})
        users_rated = d.find('a', attrs={'class':'a-size-small a-link-normal'})
        price = d.find('span', attrs={'class':'p13n-sc-price'})

        all1=[]

        if name is not None:
            #print(n[0]['alt'])
            all1.append(n[0]['alt'])
        else:
            all1.append("unknown-product")

        if rating is not None:
            #print(rating.text)
            all1.append(rating.text)
        else:
            all1.append('-1')

        if users_rated is not None:
            #print(price.text)
            all1.append(users_rated.text)
        else:
            all1.append('0')     

        if price is not None:
            #print(price.text)
            all1.append(price.text)
        else:
            all1.append('0')
        alls.append(all1)    
    return alls
```

```python
results = []
for i in range(1, no_pages+1):
    results.append(get_data(i))
flatten = lambda l: [item for sublist in l for item in sublist]
df = pd.DataFrame(flatten(results),columns=['Product Name','Rating','Total Votes', 'Price'])
df.to_csv('amazon_products.csv', index=False, encoding='utf-8')
```

##Average Weighted Recommendation

W = (Rv + Cm)/(v + m)


W = Weighted Rating

R = Average rating as of number from 1 to 5

v  = Total votes for the product

m = minimum votes required to be listed in top

C = The mean Rating across the whole report

```python
# Calculate all the components based on the  formula
v=amazon['Total Votes']
R=amazon['Rating']
C=amazon['Rating'].mean()
m=amazon['Total Votes'].quantile(0.70) 
```

```python
amazon['weighted_average']=((R*v)+ (C*m))/(v+m)

amazon.head()
```

```python
amazon_rank=amazon.sort_values('weighted_average',ascending=False)
amazon_rank.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

weight_average=amazon_rank.sort_values('weighted_average',ascending=False)
plt.figure(figsize=(12,6))
axis1=sns.barplot(x=weight_average['weighted_average'].head(10), y=weight_average['Product Name'].head(10), data=weight_average)
plt.xlim(1, 5)
plt.title('Top Products based on Ratings', weight='bold')
plt.xlabel('Weighted Average', weight='bold')
plt.ylabel('Electronic Products', weight='bold')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Using  arguments for TF-IDF
tfv = TfidfVectorizer(min_df=3,  max_features=None, 
            strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',
            ngram_range=(1, 3),
            stop_words = 'english')
```

```python
# Fitting the TF-IDF on the 'overview' text
tfv_matrix = tfv.fit_transform(amazon_rank['Product Name'])

tfv_matrix.shape
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel
# Compute the sigmoid kernel
sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
```

```python
sig[0]
```

```python
# Reverse mapping of indices and product titles
indices = pd.Series(amazon.index, index=amazon['Product Name']).drop_duplicates()
```

```python
def give_rec(title, sig=sig):
    # Get the index corresponding to product name
    idx = indices[title]

    # Get the pairwsie similarity scores 
    sig_scores = list(enumerate(sig[idx]))

    # Sort the products
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)

    # Scores of the 10 most similar product
    sig_scores = sig_scores[1:11]

    # product indices
    product_indices = [i[0] for i in sig_scores]

    # Top 10 most similar product
    return amazon_rank['Product Name'].iloc[product_indices]
```
